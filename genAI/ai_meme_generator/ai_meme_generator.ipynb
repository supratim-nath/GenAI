```python
import google.generativeai as genai
from PIL import Image
import io
import ipywidgets as widgets
from IPython.display import display, clear_output, Markdown
```

```python
# Configure your Gemini API Key here
API_KEY = "AIzaSyCoJOZ1S_AxBkRlDbwDk1vuW9J8BRR9ssY"  # Get your free key at: https://aistudio.google.com/app/apikey
genai.configure(api_key=API_KEY)

# Initialize the model
model = genai.GenerativeModel('gemini-2.5-flash')
```

```python
# --- Create the Interactive UI ---

# 1. Upload Widget
upload_widget = widgets.FileUpload(
    accept='image/*',  
    multiple=False,    
    description='Upload Image'
)

# 2. Humor Style Dropdown
style_widget = widgets.Dropdown(
    options=['Sarcastic', 'Dark Humor', 'Wholesome', 'Gen Z'],
    value='Sarcastic',
    description='Humor Style:',
)

# 3. Generate Button
generate_btn = widgets.Button(
    description='Generate Captions',
    button_style='primary',
    tooltip='Click to generate meme captions',
    icon='magic'
)

# 4. Output Area for images and text
output_area = widgets.Output()

def on_generate_clicked(b):
    with output_area:
        clear_output()
        
        if not upload_widget.value:
            print("⚠️ Please upload an image first by clicking the 'Upload Image' button.")
            return
            
        if API_KEY == "YOUR_API_KEY" or not API_KEY:
            print("⚠️ Please set your API_KEY in the cell above and re-run it.")
            return
        
        print("🤖 AI is thinking of funny captions...")
        
        try:
            # Extract image from the upload widget (handles both ipywidgets v7 and v8)
            if isinstance(upload_widget.value, tuple):
                uploaded_file = upload_widget.value[0]
            else:
                uploaded_file = list(upload_widget.value.values())[0]
                
            image_content = uploaded_file['content']
            image = Image.open(io.BytesIO(image_content))
            
            # Define the exact prompt
            prompt = f"Analyze this image and generate 5 short meme captions suitable for social media. Humor style: {style_widget.value}. Keep them funny and concise."
            
            # Generate captions with Gemini
            response = model.generate_content([prompt, image])
            
            # Clear the 'thinking' message and display results
            clear_output()
            
            # Display a resized version of the image so it fits nicely in the output
            display_image = image.copy()
            display_image.thumbnail((400, 400))
            display(display_image)
            
            # Display the markdown text
            display(Markdown("### ✨ Generated Meme Captions:\n" + response.text))
            
        except Exception as e:
            print(f"❌ An error occurred: {e}")

# Attach the click event to the button
generate_btn.on_click(on_generate_clicked)

# Display UI layout
display(Markdown("### Configure and Generate"))
display(widgets.VBox([upload_widget, style_widget, generate_btn, output_area]))
```
